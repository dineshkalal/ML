```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
datasets= sns.get_dataset_names()
```

```python
datasets
```

```python
# Display first few rows
print(iris.tail())
```

```python
#Pairplot (Scatter matrix)

#This helps visualize relationships between numerical features.

sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

```python
#Bxplot

#Boxplots are useful for visualizing distributions and detecting outliers.

plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()
```

```python
#Violin Plot

#A combination of boxplot and KDE plot to show distribution.

plt.figure(figsize=(8, 6))
sns.violinplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
#Swarm Plot

#Shows individual data points along with the distribution.

plt.figure(figsize=(8, 6))
sns.swarmplot(x="species", y="sepal_width", data=iris)
plt.show()
```

```python
#Heatmap (Correlation Matrix)

#Shows the correlation between different numerical features.
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(),annot=True,cmap="coolwarm",linewidths=0.5)
plt.show()
```

```python
#KDE Plot (Density Plot)

#Visualizes the density distribution of a feature.

plt.figure(figsize=(8, 6))
sns.kdeplot(data=iris, x="petal_length", hue="species", fill=True)
plt.show()
```

```python
#Histogram

plt.figure(figsize=(8, 6))
sns.histplot(iris, x="sepal_length", hue="species", kde=False, bins=10)
plt.show()
```
