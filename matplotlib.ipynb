```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sample DataFrame
data = {'x': range(1, 11),
        'y': [5, 7, 3, 8, 10, 6, 2, 4, 9, 3]}
df = pd.DataFrame(data)

# Line chart
plt.plot(df['x'], df['y'], marker='^',linestyle='dashed',linewidth=3)

plt.show()
```

```python
data = {'x': ['A', 'B', 'C', 'D', 'E'],
        'y': [15, 30, 25, 10, 20]}
df = pd.DataFrame(data)

# Bar chart
plt.bar(df['x'], df['y'])
plt.title('Bar Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
data = {'x': ['A', 'B', 'C', 'D', 'E'],
        'y': [15, 30, 25, 10, 20]}
df = pd.DataFrame(data)

# Horizontal bar chart
plt.barh(df['x'], df['y'])
plt.title('Horizontal Bar Chart')
plt.xlabel('Y-axis')
plt.ylabel('X-axis')
plt.grid(axis='x')
plt.show()
```

```python
data = {'x': range(1, 11),
        'y': [5, 7, 3, 8, 10, 6, 2, 4, 9, 3]}
df = pd.DataFrame(data)

# Scatter plot
plt.scatter(df['x'], df['y'],linewidth=15, marker=4)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# Sample DataFrame
data = {'labels': ['A', 'B', 'C', 'D', 'E'],
        'sizes': [15, 30, 25, 10, 20]}
df = pd.DataFrame(data)

# Pie chart
plt.pie(df['sizes'], labels=df['labels'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Sample DataFrame
import numpy as np
data = {'values': np.random.randn(1000)}
df = pd.DataFrame(data)

# Histogram
plt.hist(df['values'], bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
